### Get Lyrics Using Genius API

As stated before, the actual lyrics are not contained in the WASABI dataset so I will need to retrieve them myself. One way to do this use using [**lyricsgenius**](https://lyricsgenius.readthedocs.io/en/master/) library to that uses the [Genius API](https://docs.genius.com/#/getting-started-h1). 

```python
genius_acccess_token = 'imcqwREBVGOcGbtuu5Dp9NRtJRQAEhHZRCOheHJ7gYhMpSWUhmtHbqyuqnNQn0an'
```

```python
genius = lg.Genius(genius_acccess_token)
```

```python
artists = list(data['artist'])
titles = list(data['title'])
lyrics = []
errors = []
```

```python
# get the lyrics for each song in the dataset
start_time = time.time()
for i in range(len(artists)):
    try:
        print('\n')
        print(i)
        if ':' in titles[i]:
            titles[i] = titles[i].rstrip().split(':',1)[1]
        song = genius.search_song(title=titles[i], artist=artists[i])
        lyrics.append(song.lyrics)
    except:
        print('Cannot process for:', artists[i], titles[i])
        print('Appending empty string!')
        lyrics.append("")
        errors.append(i)
end_time = time.time()
print("Execution time: %s min" % ((end_time - start_time)/60))
```

```python
data['lyrics'] = lyrics
```
